<h1> <center> Fonction de test de <u> toutes les séries temporelles </u> avec des valeurs différentes de tension </center> </h1>

```python


def tension_test_allts(h, tension_list, u):
    
    """
    Teste une liste de valeurs pour tension (tension (N) dans yaml ou H[N] dans List*.txt) avec des 
    valeurs fixés pour h et u
    
    Paramètres
    ----------
    h : Valeur unique de h à tester en combinaison des différentes valeurs de tension
    tension_list : Liste des valeurs de tension à tester
    u : Valeur unique de u à tester en combinaison des différentes valeurs de tension
    

    """


    if __name__ == '__main__':
        for idx, row in data_list.iterrows():
            set_params = data_list.iloc[idx,:]
            ref = pd.read_csv("../data/csv/graph{}.csv".format(set_params["nc"]))

            print("""
            ####################################################################################### 
            #                                      Serie n: {}                                     #
            #######################################################################################
            """.format(set_params["nc"])
            )        



            for tension_value in tension_list: 

                print("################ Tension value: {} ################".format(tension_value))
                cfg = yaml.safe_load(open('../data/config/example.in.yaml', 'r'))

                
                cfg["cable"]["h"] = float(h)

                cfg["simulation"]["tf"] = float(set_params["tf[s]"])

                tension_value = float(tension_value)
                cfg["cable"]["tension"] = tension_value

                cfg["wakeosc"]["u"] = u


                cfg["simulation"]["dt"] = cfg["simulation"]["tf"] / len(ref) # MODIF DT
                cfg["simulation"]["dr"] = cfg["simulation"]["tf"] / len(ref) # MODIF DR


                print("Tension value : ", cfg["cable"]["tension"], " u value: ", cfg["wakeosc"]["u"],
                      " h value: ", cfg["cable"]["h"])

                dfy, _ = simulation.run_cable_wakeosc(cfg)



                plt.figure(figsize = (20,5))
                plt.plot(ref['time'], ref['y/d'], label = "ref")
                plt.plot(dfy.index, dfy['s=0.250']/0.025, label = "sim")
                plt.xlabel('time (s)')
                plt.ylabel('y/d')

                plt.legend()
                plt.show() 

                print("R2 : ",r2_score(ref['y/d'],(dfy['s=0.250']/0.025).values[:-1]))
                print("MSE : ",mean_squared_error(ref['y/d'],(dfy['s=0.250']/0.025).values[:-1])) # Moi qui ai testé
```

```python
tension_list = [32100]
h = 100
u = 2.15

tension_test_allts(h,tension_list,u)
```

<h1> <center> Test d'une <u> unique série </u> temporelle avec des valeurs différentes de tension et Traitement des résultats pour modélisation</center> </h1>

```python
tension_list = [32100]
h = 100
u = 2.15
set_params = data_list.iloc[0,:]

ref = pd.read_csv("../data/csv/graph{}.csv".format(set_params["nc"]))

for tension_value in tension_list: 

    print("################ Tension value: {} ################".format(tension_value))
    cfg = yaml.safe_load(open('../data/config/example.in.yaml', 'r'))

    
    cfg["cable"]["h"] = float(h)

    cfg["simulation"]["tf"] = float(set_params["tf[s]"])

    cfg["cable"]["tension"] = float(tension_value)

    cfg["wakeosc"]["u"] = u

    cfg["simulation"]["dt"] = cfg["simulation"]["tf"] / len(ref) # MODIF DT
    cfg["simulation"]["dr"] = cfg["simulation"]["tf"] / len(ref) # MODIF DR



    print("h value: ", cfg["cable"]["h"], " u value: ", cfg["wakeosc"]["u"]," tension value: ",cfg["cable"]["tension"])

    dfy, _ = simulation.run_cable_wakeosc(cfg)



    plt.figure(figsize = (20,5))
    plt.plot(ref['time'], ref['y/d'], label = "ref")
    plt.plot(dfy.index, dfy['s=0.250']/0.025, label = "sim")
    plt.xlabel('time (s)')
    plt.ylabel('y/d')

    plt.legend()
    plt.show() 



    print("R2 : ",r2_score(ref['y/d'],(dfy['s=0.250']/0.025).values[:-1]))
    print("MSE : ",mean_squared_error(ref['y/d'],(dfy['s=0.250']/0.025).values[:-1])) # Correct


```
